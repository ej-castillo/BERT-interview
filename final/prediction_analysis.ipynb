```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
PREDICTION_DATA = './output_csv/deep_speech/keywords_-_general/predictions-ds-g-top20.csv'
TRUE_DATA = './output_csv/deep_speech/keywords_-_general/raw_labels.csv'
```

```python
pred = pd.read_csv(PREDICTION_DATA).sort_values(by=['id'])
true = pd.read_csv(TRUE_DATA).sort_values(by=['id'])
```

```python
true.head()
```

```python
pred.head()
```

```python
analysis = pred.copy()
```

```python
df = analysis.set_index('id')
df = pd.DataFrame(df.columns.values[np.argsort(-df.values, axis=1)[:, :5]], 
                  index=df.index,
                  columns = ['1st Max','2nd Max','3rd Max', '4th Max', '5th Max']).reset_index()
df['Labels'] = df[df.columns[1:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
```

```python
df.head()
```

```python
analysis['Top Labels'] = df['Labels'].to_numpy()
analysis.head()
```

```python
analysis['Average Tag Prediction'] = analysis.mean(axis=1)
analysis.head()
```

```python
cols = pred.columns.tolist()
true_cropped = true[cols]
```

```python
true_cropped.shape
```

```python
def acc(true, pred):
    acc = 0
    for i in range(len(pred)):
        acc += abs(pred[i] - true[i])
    return 1 - (acc / len(pred))
```

```python
>>> from sklearn.metrics import accuracy_score
accuracy = []

for index, row in true_cropped.iterrows(): 
    true_labels = np.array(row.values.tolist()[1:])
    pred_labels = np.array(pred.loc[pred['id'] == row['id']].values.tolist()[0][1:])
    if index == 1:
        print(len(true_labels), len(pred_labels))
#     accuracy.append(accuracy_score(true_labels, pred_labels, normalize=False))
    accuracy.append(acc(true_labels, pred_labels))
```

```python
analysis['Accuracy'] = accuracy
```
