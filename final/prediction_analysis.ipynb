```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
PREDICTION_DATA = './output_csv/google_transcript/keywords_-_general/predictions-top20.csv'
TRUE_DATA = './output_csv/google_transcript/keywords_-_general/raw_labels.csv'
```

```python
pred = pd.read_csv(PREDICTION_DATA).sort_values(by=['id']).reset_index(drop=True)
true = pd.read_csv(TRUE_DATA).sort_values(by=['id']).reset_index(drop=True)
```

```python
pred.head()
```

```python
analysis = pred.copy()
```

```python
df = analysis.set_index('id')
df = pd.DataFrame(df.columns.values[np.argsort(-df.values, axis=1)[:, :5]], 
                  index=df.index,
                  columns = ['1st Max','2nd Max','3rd Max', '4th Max', '5th Max']).reset_index()
df['Labels'] = df[df.columns[1:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
```

```python
df.head()
```

```python
analysis['Top Labels'] = df['Labels'].to_numpy()
analysis.head()
```

```python
cols = pred.columns.tolist()
true_cropped = true[cols]
```

```python
true_cropped.head()
```

```python
def avg(true, pred):
    avg = 0
    for i in range(len(pred)):
        avg += abs(pred[i] - true[i])
    return 1 - (avg / len(pred))
```

```python
def acc(tups):
    acc = 0
    for tup in tups:
        acc += tup[1]
    return acc/len(tups)
```

```python
avgs = []
accuracies = [[] for i in range(20)]
# labels = true.columns.tolist()[2:]

# Do numerical analysis on each transcript
for index, row in true_cropped.iterrows(): 
    
    true_labels = np.array(row.values.tolist()[1:])
    pred_labels = np.array(pred.loc[pred['id'] == row['id']].values.tolist()[0][1:])
    
    labels_sorted = []
    for i, label in enumerate(pred_labels):
        labels_sorted.append((label, true_labels[i]))
    labels_sorted.sort(reverse=True)
    
    
    assert len(true_labels) == 20
    for i in range(len(true_labels)):
        accuracies[i].append(acc(labels_sorted[:i + 1]))
    
    if index == 1:
        print(true_labels)
        print(pred_labels)
        print(labels_sorted)
    
    avgs.append(avg(true_labels, pred_labels))
```

```python
analysis['Average True-Prediction Difference'] = avgs
```

```python
analysis.head()
```

```python
for i in range(20):
    row_name = 'Top ' + str(i + 1) + ' Labels Prediction Accuracy'
    analysis[row_name] = accuracies[i]
```

```python
analysis.to_csv('./output_csv/google_transcript/keywords_-_general/predictions-analysis.csv', index=False)
```
